```python
Dow_Jones = Dow_Jones.rename(columns={'Дата': 'date', 'Цена': 'price', 'Объём': 'volume', 'Изм. %': 'change %'})
NASDAQ = NASDAQ.rename(columns={'Дата': 'date', 'Цена': 'price', 'Объём': 'volume', 'Изм. %': 'change %'})
Gold = Gold.rename(columns={'Дата': 'date', 'Цена': 'price', 'Объём': 'volume', 'Изм. %': 'change %'})
Bitcoin = Bitcoin.rename(columns={'Дата': 'date', 'Цена': 'price', 'Объём': 'volume', 'Изм. %': 'change %'})
Ethereum = Ethereum.rename(columns={'Дата': 'date', 'Цена': 'price', 'Объём': 'volume', 'Изм. %': 'change %'})
Tether = Tether.rename(columns={'Date': 'date', 'Price': 'price', 'Vol.': 'volume', 'Change %': 'change %'})
```

#### 1.5 Чистим DataFrame от пустых значений

```python
Dow_Jones = Dow_Jones.dropna()
NASDAQ = NASDAQ.dropna()
Gold = Gold.dropna()
Bitcoin = Bitcoin.dropna()
Ethereum = Ethereum.dropna()
Tether = Tether.dropna()
```

#### 1.6 удаление знака "%" из столбца "Change %"

```python
Dow_Jones['change %'] = Dow_Jones['change %'].str.replace('%', '')
NASDAQ['change %'] = NASDAQ['change %'].str.replace('%', '')
Gold['change %'] = Gold['change %'].str.replace('%', '')
Bitcoin['change %'] = Bitcoin['change %'].str.replace('%', '')
Ethereum['change %'] = Ethereum['change %'].str.replace('%', '')
Tether['change %'] = Tether['change %'].str.replace('%', '')
```

#### 1.7 Создаем функцию для удаления знаков "M", "K", "B" из столбца "Volume" и правильное увеличение значений в зависимости от знаков.

```python
def clean_vol(volume):
    multipliers = {0: 10000000, 1: 1000000, 2: 100000}
    count = volume.count('.')
    if count in multipliers:
        multiplier = multipliers[count]
        if 'B' in volume:
            return int(float(volume.replace('B', '').replace(',' , '')) * multiplier)
        elif 'M' in volume:
            return int(float(volume.replace('M', '').replace(',' , '')) * (multiplier / 1000))
        elif 'K' in volume:
            return int(float(volume.replace('K', '').replace(',' , '')) * (multiplier / 1000000))
    else:
        return volume
```

#### 1.8 Применяем написанную функцию 

```python
Dow_Jones['volume'] = Dow_Jones['volume'].apply(clean_vol)
NASDAQ['volume'] = NASDAQ['volume'].apply(clean_vol)
Gold['volume'] = Gold['volume'].apply(clean_vol)
Bitcoin['volume'] = Bitcoin['volume'].apply(clean_vol)
Ethereum['volume'] = Ethereum['volume'].apply(clean_vol)
Tether['volume'] = Tether['volume'].apply(clean_vol)
```

#### 3.1 Создание новой колонки 'days_since_start' с количеством дней с начала наблюдений для построения модели линейной регресии

```python
price['days_since_start'] = (price['date'] - price['date'].min()).dt.days
price
```

#### 3.2 Построение модели для анализа цены Bitcoin

```python
# Выделение признаков и целевой переменной
X = price[['days_since_start']]
y = price['Bitcoin']

# Создание и обучение модели
model = LinearRegression()
model.fit(X, y)

# Предсказания
y_pred = model.predict(X)

# Построение графика данных и линии линейной регрессии
plt.figure(figsize=(10, 6))
plt.scatter(price['days_since_start'], price['Bitcoin'], color='orange', label='Фактические значения', s=10)
plt.plot(price['days_since_start'], y_pred, color='red', linewidth=2, label='Линейная регрессия')
plt.xlabel('Нулевая отметка это - 11/2013, 3500 это - 11/2023')
plt.ylabel('Bitcoin')
plt.title('Линейная регрессия Bitcoin')
plt.legend()
plt.show()

# Предполагая, что 2025 год соответствует определенному количеству дней с начала 11/2013
days_since_2025 = 4200

# Используем обученную модель, чтобы предсказать стоимость на 2025 год
predicted_price_2025 = model.predict([[days_since_2025]])

# Выводим предсказанную стоимость
print("Предсказанная стоимость Bitcoin на 2025 год:", predicted_price_2025)

# Вычислим процентную разницу между ценой актива в настоящем и той которую предсказывает наша модель
actual_price = price['Bitcoin'].head(1).values[0]
percentage_difference = ((predicted_price_2025 - actual_price) / actual_price) * 100

# Выводим процентное значение разницы
print('percentage_difference:', percentage_difference)
```
